```python
#DESAFIO 1
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar los datos para incluir solo las categoras deseadas
categorias_deseadas = ['Lote', 'Local', 'Bodega']
inmuebles_filtrados = inmuebles[inmuebles['Tipo'].isin(categorias_deseadas)]

plt.figure(figsize=(10, 6))
grafica = sns.histplot(data=inmuebles_filtrados,
                       x='Precio_Millon',
                       kde=True,
                       hue='Tipo',
                       binwidth=400,
                       color='red',
                       alpha=0.3,
                       edgecolor='black',
                       legend=True,
                       common_norm=False,
                       )
grafica.set_xlabel('Precio en Millones de Pesos', fontsize=14)
grafica.set_ylabel('Frecuencia', fontsize=14)
grafica.set_title('Distribucin de valores de los inmuebles en Bogot', fontsize=16)
grafica.legend(title='Tipos de Inmuebles', labels=['Lote', 'Local', 'Bodega'])
plt.text(2000, 10, 'Mximo en Lote', fontsize=12, color='blue')
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(range(0, 5001, 400))
plt.yticks(range(0, 13, 1))
plt.text(2400, 7, 'Conclusin1:\nSi queremos presupuestar la compra \nde una Bodega sera al rededor \nde 1600 millones de pesos', fontsize=12, color='black')



plt.xlim((50, 5000))  # lmites segn tus datos
plt.ylim((0, 12))      # lmites segn tus datos

# la ruta es vlida y accesible en Google Colab
ruta_guardado = '/content/drive/MyDrive/Colab Notebooks/archivos/valor_inmuebles4.png'
plt.savefig(ruta_guardado, format='png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Supongamos que tienes un DataFrame "inmuebles" con las columnas "Barrio" y "Precio_por_Metro_Cuadrado"

plt.figure(figsize=(12, 8))
sns.barplot(x='Precio_por_Metro_Cuadrado', y='Barrio', data=inmuebles, ci=None)

plt.xlabel('Precio por Metro Cuadrado')
plt.ylabel('Barrio')
plt.title('Precio por Metro Cuadrado en Diferentes Barrios')
plt.grid(axis='x')  # Agregar una cuadrcula en el eje x para facilitar la lectura

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Supongamos que tienes un DataFrame "inmuebles" con las columnas "Barrio" y "Precio_por_Metro_Cuadrado"

# Ordenar el DataFrame por "Precio_por_Metro_Cuadrado" en orden descendente
inmuebles_ordenados = inmuebles.sort_values(by='Precio_por_Metro_Cuadrado', ascending=False)
inmuebles_ordenados.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Supongamos que tienes un DataFrame "inmuebles" con las columnas "Barrio" y "Precio_por_Metro_Cuadrado"

# Calcular el promedio de "Precio_por_Metro_Cuadrado" para cada barrio
promedio_por_barrio = inmuebles.groupby('Barrio')['Precio_por_Metro_Cuadrado'].mean().reset_index()
promedio_por_barrio_ordenado=promedio_por_barrio.sort_values(by='Precio_por_Metro_Cuadrado', ascending=False)
top_5_barrios=promedio_por_barrio_ordenado.head(10)

plt.figure(figsize=(12,12))
grafica=sns.barplot(x='Precio_por_Metro_Cuadrado', y='Barrio', data=top_5_barrios, errcolor='black')
plt.xlabel=('Precio por metro cuadrado en promedio')
plt.ylabel=('Barrios')
plt.title('Top 10 Barrios con el promedio ms alto de m2')
plt.grid(axis='x') # Agregar una cuadrcula en el eje x para facilitar la lectura


ruta='/content/drive/MyDrive/Colab Notebooks/archivos/top_10_barrios_mas_caros.png'
plt.savefig(ruta, format='png')
plt.show()
```
