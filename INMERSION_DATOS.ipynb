```python
import matplotlib.pyplot as plt
import seaborn as sns

# Supongamos que tienes un DataFrame "inmuebles" con las columnas "Barrio" y "Precio_por_Metro_Cuadrado"

# Ordenar el DataFrame por "Precio_por_Metro_Cuadrado" en orden descendente
inmuebles_ordenados = inmuebles.sort_values(by='Precio_por_Metro_Cuadrado', ascending=False)
inmuebles_ordenados.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Supongamos que tienes un DataFrame "inmuebles" con las columnas "Barrio" y "Precio_por_Metro_Cuadrado"

# Calcular el promedio de "Precio_por_Metro_Cuadrado" para cada barrio
promedio_por_barrio = inmuebles.groupby('Barrio')['Precio_por_Metro_Cuadrado'].mean().reset_index()
promedio_por_barrio_ordenado=promedio_por_barrio.sort_values(by='Precio_por_Metro_Cuadrado', ascending=False)
top_5_barrios=promedio_por_barrio_ordenado.head(10)

plt.figure(figsize=(12,12))
grafica=sns.barplot(x='Precio_por_Metro_Cuadrado', y='Barrio', data=top_5_barrios, errcolor='black')
plt.xlabel=('Precio por metro cuadrado en promedio')
plt.ylabel=('Barrios')
plt.title('Top 10 Barrios con el promedio más alto de m2')
plt.grid(axis='x') # Agregar una cuadrícula en el eje x para facilitar la lectura


ruta='/content/drive/MyDrive/Colab Notebooks/archivos/top_10_barrios_mas_caros.png'
plt.savefig(ruta, format='png')
plt.show()
```

##3.1.Calculo del valor del m2 por barrio

```python
inmuebles['valor_m2_millon']=inmuebles['Precio_Millon']/inmuebles['Area']
inmuebles.head(3)
```

```python
inmuebles.groupby('Barrio').mean()
```

```python
datos_barrio=inmuebles.groupby('Barrio').sum()
datos_barrio
```

```python
datos_barrio['valor_m2_barrio']=datos_barrio['Precio_Millon']/datos_barrio['Area']
datos_barrio
```

en este caso ha sumado todos los precios de un barrio y dividirlo entre la suma de todas esas áreas.

##3.2.Inclusion de la nueva variable al dataframe original

```python
m2_barrio=dict(datos_barrio['valor_m2_barrio'])
```

```python
inmuebles['valor_m2_barrio']=inmuebles['Barrio']
inmuebles['valor_m2_barrio']=inmuebles['valor_m2_barrio'].map(m2_barrio)
inmuebles.head(5)
```

```python
top_barrios=inmuebles['Barrio'].value_counts()[:10].index
top_barrios
```

```python
datos_barrio.reset_index(inplace=True)
datos_barrio #esto vuelve a barrio en una columna
```

```python
datos_barrio.query('Barrio in @top_barrios')
```
